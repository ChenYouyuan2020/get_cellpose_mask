```python
import numpy as np
import cv2

path = r"data/cellpose_masks.txt"
data = np.loadtxt(path, delimiter=',')
print(data.shape)
```

```python
# 对细胞进行分组
data = data %10 
```

```python
import matplotlib.pyplot as plt
img = cv2.normalize(data, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U, mask=None)
plt.figure()
plt.imshow(img[4000:6000, 4000:6000])
cv2.imwrite('output_image.jpg', img) # 保存图片为jpg 格式
```
